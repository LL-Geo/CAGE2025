# Drift correciton for our field data


Load our tools (very important)

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from pyproj import Proj
import netCDF4 as nc
from scipy.interpolate import interp1d
from datetime import datetime
import boule as bl
```

```python
names = ['LINE', 'STATION', 'ALT.', 'GRAV.', 'SD.', 'TILTX', 'TILTY', 'TEMP', 'TIDE', 'DUR', 'REJ', 'TIME', 'DEC.TIME+DATE', 'TERRAIN', 'DATE']
Data = pd.read_csv('./filed_data/GRAV.txt', skiprows=22, names=names, delimiter=r'\s+')

# Read GPS data
GPS=pd.read_csv('./filed_data/GGG011025.csv')

# Match our graivty measurement with GPS
Data = pd.merge(Data, GPS, on = ['STATION', 'LINE'], how='left')
```

```python
Data
```

```python
Data_1st_day = Data[Data.DATE=='2025/10/01']
Data_2nd_day = Data[Data.DATE=='2025/10/02']
```

```python
plt.scatter(Data.STATION, Data.LINE, c=Data['GRAV.'], cmap='viridis')
plt.xlim(0,30)
plt.ylim(150,550)
plt.xlabel('Station')
plt.ylabel('Line')
plt.title('Raw Gravity')
plt.colorbar(label='mGal')  # Adds a color scale legend
plt.gca().invert_xaxis()
```

```python
plt.scatter(Data.Lon, Data.Lat, c=Data['GRAV.'], cmap='viridis')

plt.xlabel('Lon')
plt.ylabel('Lat')
plt.title('Raw Gravity')
plt.colorbar(label='mGal')  # Adds a color scale legend
```

```python
line_number=200 #You could change number in here to look different line

plt.figure(figsize=(8,2))
plt.scatter(Data.STATION[Data.LINE==line_number],Data['GRAV.'][Data.LINE==line_number])
even_ticks = np.arange(2, 30, 2)
plt.xticks(even_ticks)

plt.xlabel('Station')
plt.ylabel('Raw Gravity(mGal)')
plt.title('Line %d' %line_number)
plt.gca().invert_xaxis()
```

```python
Data_1st_day
```

```python
Data_1st_day = Data_1st_day.assign(DATE=pd.to_datetime(Data['DATE'], format='%Y/%m/%d'))
Data_1st_day = Data_1st_day.assign(DateTime=pd.to_datetime(Data_1st_day['DATE'].dt.strftime('%Y-%m-%d') 
                                                       + ' ' + Data_1st_day['TIME'], format='%Y-%m-%d %H:%M:%S'))

Data_1st_day.loc[:,'DateTime'] = pd.to_datetime(Data_1st_day['DateTime'])

datetime_values = Data_1st_day['DateTime'].copy()
datetime_strings = [str(dt) for dt in datetime_values]

reference_time = min(datetime.strptime(dt, '%Y-%m-%d %H:%M:%S') for dt in datetime_strings)
seconds_elapsed = [(datetime.strptime(dt, '%Y-%m-%d %H:%M:%S') - reference_time).total_seconds() for dt in datetime_strings]
Data_1st_day.loc[:,'seconds_elapsed'] = seconds_elapsed
Data_control_1 = Data_1st_day[(Data_1st_day.STATION==100) & (Data_1st_day.LINE==0)]
```

```python
Data_2nd_day = Data_2nd_day.assign(DATE=pd.to_datetime(Data['DATE'], format='%Y/%m/%d'))
Data_2nd_day = Data_2nd_day.assign(DateTime=pd.to_datetime(Data_2nd_day['DATE'].dt.strftime('%Y-%m-%d') 
                                                       + ' ' + Data_2nd_day['TIME'], format='%Y-%m-%d %H:%M:%S'))

Data_2nd_day.loc[:,'DateTime'] = pd.to_datetime(Data_2nd_day['DateTime'])

datetime_values = Data_2nd_day['DateTime'].copy()
datetime_strings = [str(dt) for dt in datetime_values]

reference_time = min(datetime.strptime(dt, '%Y-%m-%d %H:%M:%S') for dt in datetime_strings)
seconds_elapsed = [(datetime.strptime(dt, '%Y-%m-%d %H:%M:%S') - reference_time).total_seconds() for dt in datetime_strings]
Data_2nd_day.loc[:,'seconds_elapsed'] = seconds_elapsed
Data_control_2 = Data_2nd_day[(Data_2nd_day.STATION==100) & (Data_2nd_day.LINE==0)]
```

```python
df_1 = Data_1st_day[['STATION','seconds_elapsed' ,'GRAV.']]
mean_Data_1 = df_1.groupby('STATION').mean() # take mean of two measurements at each location
mean_Data_1.reset_index(inplace=True)

df_2 = Data_2nd_day[['STATION','seconds_elapsed' ,'GRAV.']]
mean_Data_2 = df_2.groupby('STATION').mean() # take mean of two measurements at each location
mean_Data_2.reset_index(inplace=True)
```

```python
# That's our gravity drift

x_1 = np.array(Data_control_1['seconds_elapsed'])
y_1 = Data_control_1['GRAV.']
m_1, b_1 = np.polyfit(x_1, y_1, 1) 
slope_label_1 = f" Drift_1: \n Slope: {m_1:.1e} \n b: {b_1:.2f}"

Drift_1 = np.vstack((x_1, m_1*x_1 + b_1)).T


x_2 = np.array(Data_control_2['seconds_elapsed'])
y_2 = Data_control_2['GRAV.']
m_2, b_2 = np.polyfit(x_2, y_2, 1) 
slope_label_2 = f" Drift_2: \n Slope: {m_2:.1e} \n b: {b_2:.2f}"

Drift_2 = np.vstack((x_2, m_2*x_2 + b_2)).T



plt.figure(figsize=(10,2))
plt.subplot(1,2,1)
plt.scatter(x_1, y_1)
plt.plot(x_1, m_1*x_1 + b_1, label=slope_label_1)
plt.legend()
plt.xlabel('Seconds Elapsed (s)')
plt.ylabel('Gravity Drift (mGal)')

plt.subplot(1,2,2)
plt.scatter(x_2, y_2)
plt.plot(x_2, m_2*x_2 + b_2, label=slope_label_2)
plt.legend()
plt.xlabel('Seconds Elapsed (s)')
plt.ylabel('Gravity Drift (mGal)')
```

```python
# group by station (taking the mean here)
df_1 = Data_1st_day[['STATION', 'LINE','seconds_elapsed','GRAV.', 'Lon', 'Lat', 'Height_Ellipsoid_m']]
df_1 = df_1[df_1['LINE']!=0]
mean_Data_1 = df_1.groupby(['STATION','LINE']).mean() # take mean of three measurements at each location
mean_Data_1.reset_index(inplace=True)

df_2 = Data_2nd_day[['STATION','LINE', 'seconds_elapsed','GRAV.', 'Lon', 'Lat', 'Height_Ellipsoid_m']]
df_2 = df_2[df_2['LINE']!=0]
mean_Data_2 = df_2.groupby(['STATION','LINE']).mean() # take mean of three measurements at each location
mean_Data_2.reset_index(inplace=True)
```

```python
# Let's look at out data

x_new_1 = mean_Data_1['seconds_elapsed']
f_1 = interp1d(Drift_1[:,0],Drift_1[:,1], kind='linear')
drift_1 = f_1(x_new_1)

x_new_2 = mean_Data_2['seconds_elapsed']
f_2 = interp1d(Drift_2[:,0],Drift_2[:,1], kind='linear')
drift_2 = f_2(x_new_2)

plt.figure(figsize=(10,2))
plt.subplot(1,2,1)
plt.plot(Drift_1[:,0],Drift_1[:,1])
plt.scatter(mean_Data_1['seconds_elapsed'], mean_Data_1['GRAV.'])
plt.plot(x_new_1,drift_1, linestyle='--')
plt.xlim(min(mean_Data_1['seconds_elapsed']), max(mean_Data_1['seconds_elapsed']))
plt.xlabel('Seconds Elapsed (s)')
plt.ylabel('Gravity (mGal)')
plt.title('Day 1')

plt.subplot(1,2,2)
plt.plot(Drift_2[:,0],Drift_2[:,1])
plt.scatter(mean_Data_2['seconds_elapsed'], mean_Data_2['GRAV.'])
plt.plot(x_new_2,drift_2, linestyle='--')
plt.xlim(min(mean_Data_2['seconds_elapsed']), max(mean_Data_2['seconds_elapsed']))
plt.xlabel('Seconds Elapsed (s)')
plt.ylabel('Gravity (mGal)')
plt.title('Day 2')
```

```python
drift_1
```

```python
drift_corrected_1 = mean_Data_1['GRAV.'] - drift_1
drift_corrected_2 = mean_Data_2['GRAV.'] - drift_2

mean_Data_1['grav_drift'] = drift_corrected_1
mean_Data_2['grav_drift'] = drift_corrected_2

#Save to file
combined = pd.concat([mean_Data_1,mean_Data_2])
combined.to_csv('Processed_Data.csv')
```

```python
line_number=200 #You could change number in here to look different line

plt.figure(figsize=(8,2))
plt.scatter(combined.STATION[combined.LINE==line_number],combined['grav_drift'][combined.LINE==line_number])
even_ticks = np.arange(2, 30, 2)
plt.xticks(even_ticks)

plt.xlabel('Station')
plt.ylabel('Drift corrected Gravity(mGal)')
plt.title('Line %d' %line_number)
plt.gca().invert_xaxis()
```

```python
line_number=200 #You could change number in here to look different line

plt.figure(figsize=(8,2))
plt.scatter(combined.STATION[combined.LINE==line_number],combined['Height_Ellipsoid_m'][combined.LINE==line_number])
even_ticks = np.arange(2, 30, 2)
plt.xticks(even_ticks)

plt.xlabel('Station')
plt.ylabel('Elevation (m)')
plt.title('Line %d' %line_number)
plt.gca().invert_xaxis()
```
